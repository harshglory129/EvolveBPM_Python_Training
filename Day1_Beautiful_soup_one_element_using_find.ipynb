```python
# Necessary Imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Gets source from the site using get function
mycode = requests.get('https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops').text
```

```python
mycode
```

```python
print(type(mycode))
```

```python
#html_parser
soup_data = BeautifulSoup(mycode,'lxml')
```

```python
print(type(soup_data))
```

```python
mydiv1 = soup_data.find('div',class_='product-wrapper card-body')
```

```python
print(mydiv1)
```

```python
product_title = mydiv1.find('a',class_='title')['title']
```

```python
print(product_title)
```

```python
product_link = mydiv1.find('a',class_='title')['href']
```

```python
product_link
```

```python
product_cost = mydiv1.find('h4',class_='price float-end card-title pull-right').text
```

```python
product_cost
```

```python
description = mydiv1.find('p', class_='description card-text').text
```

```python
description
```

```python
star_ratings = mydiv1.find_all('p')[2]['data-rating']
```

```python
print(star_ratings)
```

```python
product_review = mydiv1.find('p', class_='review-count float-end').text
```

```python
product_review
```

```python
first_product = {
    'Product Title':product_title,
    'Product Link' :product_link,
    'Product Price':product_cost,
    'Product Desc':description,
    'Product Rating':star_ratings,
    'Product Review' :product_review
}
```

```python
df = pd.DataFrame([first_product])
```
